```
"""
Created on Sat Aug 31 14:39:14 2019

@author: haseena
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#importing important libraries first
from sklearn import svm,datasets
iris = pd.read_csv('iris.csv')
```

```
#checking the dataset
iris.head()
```

```
#creating pairplot to visualise similarities and especially difference btw the species
sns.pairplot(data = iris, hue = 'variety',palette = 'Set2')
```

```
#Train test split
from sklearn.model_selection import train_test_split

```

```
#seperating the independent variables from dependent variables
x = iris.iloc[:,:-1]
y = iris.iloc[:,4]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```
#importing support vector
from sklearn.svm import SVC
```

```
#Training and fitting the model
svm.SVC(kernel = 'linear', C = 1, gamma = 0.1).fit(x,y)
model = SVC()
model.fit(x_train, y_train)
```

```

#Prediction from the trained model
pred = model.predict(x_test)
```

```
#model evaluation
#importing classification report and confusion matrix

from sklearn.metrics import classification_report, confusion_matrix

#confusion matrix has how many iris data are correctly classified and misclassified
#true positive & true negative
#false positive & false negative
```

```
conf = confusion_matrix(y_test, pred)
rep = classification_report(y_test, pred)

print(pred)
print(conf)
print(rep)
```
