```
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Aug 31 11:30:05 2019

@author: haseena
"""
#common imports
import numpy as np
import os
import matplotlib


np.random.seed(314)

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
from sklearn.datasets import load_wine
from IPython.display import SVG
from graphviz import Source
from IPython.display import display

#load dataset
data = load_wine()

#feature matrix
X = data.data
print('shape of X is',X.shape)

#target vector
y = data.target
#class lebels
labels = data.feature_names
print(labels)
#print dataset discription
print(data.DESCR)
#Training a decision tree
#Now we build Decision tree classifier and fit the model
estimator = DecisionTreeClassifier()
estimator.fit(X,y)
#Visualize the tree using graphviz
graph = Source(tree.export_graphviz(estimator,
                                    out_file = None,
                                    feature_names = labels,class_names = ['0','1','2'],
                                    filled = True))

display(SVG(graph.pipe(format = 'svg')))

#Predicting classes and class probabilities

new = [[1.32e+01, 1.78e+00, 2.14e+00, 1.12e+01, 1.00e+02, 2.65e+00, 2.76e+00,
        2.60e-01, 1.28e+00, 4.38e+00, 1.05e+00, 3.40e+00, 1.05e+03]]
#probability for each class [class0,class1,class2]
print(estimator.predict_proba(new))
#predicting the exact class to which it is more probable to belong
print(estimator.predict(new))
```
